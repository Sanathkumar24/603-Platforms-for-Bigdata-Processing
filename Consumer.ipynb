```python
from kafka import KafkaConsumer                   #importing kafka consumer library
import time
import json
```

```python
kafka_consumer = KafkaConsumer('Assignment',
                         bootstrap_servers=['localhost:9092'],consumer_timeout_ms = 10000,
                           value_deserializer=lambda m: json.loads(m.decode('utf-8')))     #assingning kafka topic to consumer
```

```python
lat,long=[],[]
```

```python
for m in kafka_consumer:
#     sleep(5)
    print(m.value)
    json_fied = m.value
    lat.append(json_fied['iss_position']['latitude'])             #loading through the data and creating a list
    long.append(json_fied['iss_position']['longitude'])
    
```

```python
len(lat),len(long)
```

```python
with open('lat.txt', 'w') as filehandle:
    for listitem in lat:
        filehandle.write(f'{listitem}\n')
```

```python
with open('long.txt', 'w') as filehandle:
    for listitem in long:
        filehandle.write(f'{listitem}\n')
```
