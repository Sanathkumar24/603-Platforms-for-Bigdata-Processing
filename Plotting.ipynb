```python
import pandas
import geopandas as gpd
import matplotlib.pyplot as plt
    
```

```python
lat,long = [],[]
```

```python
with open('long.txt','r') as longitude:
    for line in longitude:
        long.append(line[:-1])

with open('lat.txt','r') as latitude:
    for line in latitude:
        lat.append(line[:-1])    
```

```python
p = pandas.DataFrame(zip(lat,long),columns=['Latitude','Longitude'])
p['latitude'] = p['Latitude'].astype(float)
p['longitude'] = p['Longitude'].astype(float)
```

```python
worldmap = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
fig, ax = plt.subplots(figsize=(20, 10))
worldmap.plot(color="lightgrey", ax=ax)

# Plotting our Impact Energy data with a color map
x = p['longitude']
y = p['latitude']
# z = df['Impact Energy [kt]']
plt.scatter(x, y, alpha=0.6)
```
